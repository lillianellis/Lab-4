**FIRST DATA SET: BOING BOING CANDY HIERARCHY 2016 SURVEY RESPONSES:**

**Introduction:**

The data set "BOING-BOING-CANDY-HIERARCHY-2016-SURVEY-Responses" is a survey released in 2016 with a goal to find information on the "best and worst treats of 2016." The link to the article that uses the data: https://boingboing.net/2016/10/31/the-candy-hierarchy-for-2016.html. This survey has been relased throughout the years. The 2015 survey, very similar to the 2016 one, was based on 518,605 data points from 5,459 individuals. 

**Dataset Preparation**

First, I read the excel file into the data frame structure, and printed the first 5 rows with the head method to see what the data looked like when it was first read in.

```python
#one line to open and read a excel filed into a data frame data structure
df3 = pd.read_excel('BOING-BOING-CANDY-HIERARCHY-2016-SURVEY-Responses.xlsx')

#print the first few lines of the data fram, default is 5 lines
df3.head()
```

I first noticed that the headers were extremley messy in that they were long, hard to read, and complicated. I renamed the column headers. 

```python
#fixing the headers, the headers are weirdly spaced and not lined up properly

#with the rename method, I called the header name and renamed it

#inplace directly replaces it in your list 
#this makes a permanent change, a view (not just editing the view)

df3.rename(columns = {'Timestamp':'time',
                     'Are you going actually going trick or treating yourself?' : 'trick_or_treat',
                      'Your gender:' : 'gender',
                      'How old are you?' : 'age',
                      'Which country do you live in?' : 'country',
                      'Which state, province, county do you live in?' : 'province',
                     #'[100 Grand Bar]', '100_grand_bar',
                     list(df3)[6] : '100_grand_bar'},
          inplace = True)

df3.head()
```

Next, I noticed that there were multiple instances where there were multiple answers for one value. People had filled out multiple answers for US, especially. 

To find the number of people from the US, I first turned everything to capital letters to minimalize the positibilities for words for the US and other values. 

```python
#turning into capitals with str.upper method
df3['country'] = df3['country'].str.upper()
#print header
df3.head()
```

Now that I had everything in that column capitalized, the possibilities of different combinations were more limited. I then used the .fromkeys and .replace methods to replace all USA values with "USA" and all UK values with "UK"

I also notived that MEH, JOY, and DESPAIR were long strings for a rating system. I decided to change these to numbers

```python
#defining all the posibilities of the term USA (now that we have capitalized, we have less possibilities)
usa_replace = dict.fromkeys(['US', 'UNITED STATES', 'UNITED STATES OF AMERICA', 'USSA', 'USA!', 'U.S.A', 'U.S.', 'UNITED STATE', 'USA! USA! USA!'], 'USA')
df3 = df3.replace(usa_replace)
uk_replace = dict.fromkeys(['UNITED KINGDOM', 'ENGLAND'], 'UK')
df3 = df3.replace(uk_replace)

#Next, replacing MEH, JOY, and DESPAIR with number values to clear up data set 
df3.replace(['MEH', 'JOY', 'DESPAIR'], [1,2,3])


```

**Second Data Set: Public School Attendance US**

**Introduction:** 

This data set is attendance, pulled from public schools around the United States. This data is from the 2007-2008 academic year. The data set is organized by school and state level.  

**Dataset Preparation**

First, I read in the excel file, and printed the first 5 rows with the head method to see what I was working with.

```python
#one line to open and read a csv filed into a data fram data structure
df2 = pd.read_excel('attendance.xls')

#print the first few lines of the data frame, default is 5 lines
df2.head()
```

The first thing I noticed when reading in this data set was that the title of the data set read as the first column name. Additionally, all of the column names were not read in. My first priority was to organize this by renaming each column so it corresponded with it's data. There was one number after each column name that we could not figure out the value of, so I named this column as additional of "other" info. 

```python
#naming the columns
#replacing string on left with the string on right
#I completley renamed each column to align with the data because it was read in differentley 
df2.rename(columns = {list(df2)[0]:'Location Name',
                      list(df2)[1]: 'Total ADA percent of enrollment',
                      list(df2)[2]: '(other percent enrollment info)',
                      list(df2)[3]: 'Total average hours in a school day',
                      list(df2)[4]: '(other school day info)',
                      list(df2)[5]: 'Total average days in a school year',
                      list(df2)[6]: '(other school year info)',
                      list(df2)[7]: 'Total average hours in a school year',
                      list(df2)[8]: '(other school year info)',
                      list(df2)[9]: 'Elementary School: ADA percent of enrollment',
                      list(df2)[10]: '(other percent enrollment info)',
                      list(df2)[11]: 'Elementary School: Average hours in school day',
                      list(df2)[12]: '(other school day info)',
                      list(df2)[13]: 'Secondary Schools: ADA percent of enrollment',
                      list(df2)[14]: '(other percent enrollment info)',
                      list(df2)[15]: 'Secondary School: Average hours in school day',
                      list(df2)[16]: '(other school day info)',},
          inplace = True)

df2.head()
```

I then deleted unwanted rows at beginning which were read-in incorrectly as data, but were originally headers 

```python
#the drop method drops index 0,1,2 as these were read in incorrectly (and because I have already renamed the columns
#with data)
df2.drop(df2.index[[0,1,2]], inplace = True)
df2
```

Deleting unwanted columns which are read in as empty (NaN)

Next, I noticed that alot of the rows are empty. My next step was to replace all null values with y. 

```python
#filling null values with y
df2 = df2.fillna('y')
#print tail to display replaced null values 
df2.tail()
```
